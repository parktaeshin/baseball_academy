```python
# 290명 선수들
batter_name = list(result["Batter"])
```

```python
# 올스타전, 포시기록 drop
## 올스타전 

cleansing2 = cleansing.copy()
drop_index1 = cleansing2[(cleansing2["Int_Month"] == 10) &
           (cleansing2["Int_Day"] > 1)].index
drop_index2 = cleansing2[(cleansing2["Int_Month"] == 7) &
           (cleansing2["Int_Day"] == 21)].index

cleansing2.drop(drop_index1, axis = 0, inplace = True)
cleansing2.drop(drop_index2, axis = 0, inplace = True)

```

```python
cleansing2[cleansing2["Str_Batter"] == "페르난데스"] # 645타석 잘나옴
```

```python
cleansing2[cleansing2["Str_Batter"] == "페르난데스"].groupby(["Int_Month","Int_Day"])["Str_Result"].count() # 144경기인데 143경기나옴...
```

```python
# G

for name in batter_name :
  index = result[result["Batter"] == name].index
  result.loc[index,"G"] = len(cleansing2[cleansing2["Str_Batter"] == name].groupby(["Int_Month","Int_Day"])["Str_Result"].count())
```

```python
# PA

for name in batter_name :
  index = result[result["Batter"] == name].index
  result.loc[index,"PA"] = len(cleansing2[cleansing2["Str_Batter"] == name])
```

```python
# BB

for name in batter_name :
  index = result[result["Batter"] == name].index
  temp = cleansing2[cleansing2["Str_Batter"] == name]
  result.loc[index,"BB"] = len(temp[temp["Str_Result"] == "볼넷"])
```

```python
# HBP

for name in batter_name :
  index = result[result["Batter"] == name].index
  temp = cleansing2[cleansing2["Str_Batter"] == name]
  result.loc[index,"HBP"] = len(temp[temp["Str_Result"] == "몸에맞는볼"])
```

```python
# 1B
# 1루타 : 안타, 내야안타, 번트안타
# 2루타 : 2루타
# 3루타 : 3루타
# 홈런 : 홈런, 홈런(그라운드)

for name in batter_name :
  index = result[result["Batter"] == name].index
  temp = cleansing2[cleansing2["Str_Batter"] == name]
  result.loc[index,"1B"] = len(temp[(temp["Str_Result"] == "안타") | (temp["Str_Result"] == "내야안타") | (temp["Str_Result"] == "번트안타")])
```

```python
# 2B
# 1루타 : 안타, 내야안타, 번트안타
# 2루타 : 2루타
# 3루타 : 3루타
# 홈런 : 홈런, 홈런(그라운드)

for name in batter_name :
  index = result[result["Batter"] == name].index
  temp = cleansing2[cleansing2["Str_Batter"] == name]
  result.loc[index,"2B"] = len(temp[(temp["Str_Result"] == "2루타")])
```

```python
# 3B
# 1루타 : 안타, 내야안타, 번트안타
# 2루타 : 2루타
# 3루타 : 3루타
# 홈런 : 홈런, 홈런(그라운드)

for name in batter_name :
  index = result[result["Batter"] == name].index
  temp = cleansing2[cleansing2["Str_Batter"] == name]
  result.loc[index,"3B"] = len(temp[(temp["Str_Result"] == "3루타")])
```

```python
# 홈런
# 1루타 : 안타, 내야안타, 번트안타
# 2루타 : 2루타
# 3루타 : 3루타
# 홈런 : 홈런, 홈런(그라운드)

for name in batter_name :
  index = result[result["Batter"] == name].index
  temp = cleansing2[cleansing2["Str_Batter"] == name]
  result.loc[index,"HR"] = len(temp[(temp["Str_Result"] == "홈런") | (temp["Str_Result"] == "홈런(그라운드)")])
```
